```python
def datetrick (data):
    dic = {'janeiro': '01', 'fevereiro' :'02', 'março':'03', 'abril': '04', 'maio': '05',
           'junho':'06', 'julho':'07', 'agosto':'08', 'setembro':'09', 'outubro':'10',
           'novembro':'11', 'dezembro':'12'}
    dia, mes, ano =data.split(" de ")
    return '%s/%s/%s'%(dia, dic[mes], ano)

fl['Date'] = fl['Date'].apply(datetrick)
```

```python
fl
```

```python
fl.dtypes
```

```python
from datetime import datetime
fl['Date'] = pd.to_datetime(fl['Date'], dayfirst=True)
#rain['data'] = pd.to_datetime(rain['data'], dayfirst=True)
fl.dtypes
```
