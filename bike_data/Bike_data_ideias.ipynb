```python
ax = sns.boxplot(y="Pinheiros", x="fds", data=df, orient = 'v', width=0.2)
ax.figure.set_size_inches(8, 6)
ax.set_title('Viagens Sentido Pinheiros', fontsize = 20)
ax.set_ylabel('NÂº Viagens', fontsize = 16)
ax.set_xlabel('Final de Semana', fontsize = 16)
ax
```
